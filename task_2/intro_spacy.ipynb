```python
#define some parameters  
noisy_pos_tags = ["PROP"]
min_token_length = 2

#Function to check if the token is a noise or not  
def isNoise(token):     
    is_noise = False
    if token.pos_ in noisy_pos_tags:
        is_noise = True 
    elif token.is_stop == True:
        is_noise = True
    elif len(token.text.strip()) <= min_token_length:
        is_noise = True
    return is_noise 
def cleanup(token, lower = True):
    if lower:
       token = token.lower()
    return token.strip()
```

```python
# top unigrams used in the reviews 
from collections import Counter
cleaned_list = [cleanup(word.text.strip()) for word in document if not isNoise(word)]
Counter(cleaned_list) .most_common(5)
```

```python
labels = set([w.label_ for w in document.ents]) 
for label in labels: 
    entities = [cleanup(e.text.strip(), lower=False) for e in document.ents if label==e.label_] 
    entities = list(set(entities)) 
    print (label,entities)
```

```python
# check all adjectives used with a word 
def pos_words (sentence, token, ptag):
    sentences = [sent for sent in sentence.sents if token in sent.text.strip()]     
    pwrds = []
    for sent in sentences:
        for word in sent:
            #if word in word.text.strip(): 
            pwrds.extend([child.text.strip() for child in word.children
                                                      if child.pos_ == ptag] )
    return Counter(pwrds).most_common(20)

pos_words(document, 'Standard', "NOUN")
```
